```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_metrics_distributions_by_device(df):
    # Get unique metrics
    metrics = df['metric'].unique()

    for metric in metrics:
        fig, axes = plt.subplots(1, 2, figsize=(10, 3), sharey=True)
        
        for i, state in enumerate(['open', 'closed']):
            subset = df[
                (df['metric'] == metric) &
                (df['state'] == state)
            ]
            
            if not subset.empty:
                sns.histplot(
                    data=subset,
                    x='average_value',
                    hue='device',
                    ax=axes[i],
                    kde=True,
                    element='step',
                    stat='count',
                )
                axes[i].set_title(f"{metric} - {state.capitalize()}")
                axes[i].set_xlabel("Average Value")
                axes[i].set_ylabel("Density" if i == 0 else "")
            else:
                axes[i].set_visible(False)
        
        plt.suptitle(f"Distribution of {metric} by Device and State", fontsize=14)
        plt.tight_layout()
        plt.show()
```

# older Adults

```python
older_adults_average_level = pd.read_csv("../data/processed/older_adults/average_measurements.csv")
```

### Average level

```python
plot_metrics_distributions_by_device(older_adults_average_level)
```

# Students

```python
students_average_level = pd.read_csv("../data/processed/students/average_measurements.csv")
```

### Average level

```python
plot_metrics_distributions_by_device(students_average_level)
```
