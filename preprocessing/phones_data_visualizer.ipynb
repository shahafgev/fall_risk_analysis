# How the entire process looks like

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_processing_stages(
    df_front_raw,
    delay_sec,              # if < 0, shift FRONT time by delay_sec (left); if >= 0, no effect (was applied to BACK before)
    step_time_sec,
    fs=100,
    start_offset=5.0,
    duration=25.0
):
    """
    FRONT phone only. Visualizes:
      1) Raw acc_x & acc_z
      2) FRONT after alignment by delay (time shift applied only if delay_sec < 0)
      3) Final trimmed window (acc_x & acc_z)
      4) Final sway band (0.510 Hz), acc_x & acc_z

    Expects columns: 'time', 'acc_x', 'acc_z'. Uses only butter_filter (defined elsewhere).
    """

    for col in ("time", "acc_x", "acc_z"):
        if col not in df_front_raw.columns:
            raise ValueError(f"FRONT dataframe missing column '{col}'")

    # -------------------
    # 1) FRONT raw acc_x & acc_z
    # -------------------
    fig1, ax1 = plt.subplots(1, 1, figsize=(12, 4), sharex=False)
    ax1.plot(df_front_raw['time'], df_front_raw['acc_x'], label='acc_x')
    ax1.plot(df_front_raw['time'], df_front_raw['acc_z'], label='acc_z')
    ax1.set_title("FRONT: raw accelerations")
    ax1.set_ylabel("m/s")
    ax1.set_xlabel("Time (s)")
    ax1.grid(True)
    ax1.legend()
    fig1.suptitle("Raw signals (FRONT)", y=0.98)
    plt.tight_layout()

    # -------------------
    # 2) After alignment by delay (FRONT only)
    #    Original 2-stream logic shifted BACK when delay_sec>0 and FRONT when delay_sec<0.
    #    With FRONT-only, we preserve the latter: shift FRONT only if delay_sec < 0.
    # -------------------
    df_f_al = df_front_raw.copy()
    if delay_sec < 0:
        df_f_al['time'] = df_f_al['time'] + delay_sec  # shift left if FRONT was lagging
    # if delay_sec >= 0, no change (since it previously applied to BACK)

    fig2, ax2 = plt.subplots(1, 1, figsize=(12, 4), sharex=True)
    ax2.plot(df_f_al['time'], df_f_al['acc_x'], label='acc_x')
    ax2.plot(df_f_al['time'], df_f_al['acc_z'], label='acc_z')
    ax2.set_title(f"FRONT aligned by delay ({delay_sec:.2f}s)")
    ax2.set_xlabel("Time (s)")
    ax2.set_ylabel("m/s")
    ax2.grid(True)
    ax2.legend()
    fig2.suptitle("Signals after alignment (FRONT)", y=0.98)
    plt.tight_layout()

    # -------------------
    # 3) Final trimmed window (FRONT acc_x & acc_z)
    # -------------------
    if step_time_sec is None:
        print("step_time_sec is None  skipping trimmed window & sway band.")
        return

    start = float(step_time_sec) + float(start_offset)
    end   = start + float(duration)

    t = df_f_al['time'].to_numpy()
    i0 = np.searchsorted(t, start, side='left')
    i1 = np.searchsorted(t, end,   side='left')

    if i0 >= len(t) or i1 <= i0:
        print("Trim window outside available time span.")
        return

    df_f_win = df_f_al.iloc[i0:i1].reset_index(drop=True)

    fig3, ax3 = plt.subplots(1, 1, figsize=(12, 4), sharex=True)
    ax3.plot(df_f_win['time'], df_f_win['acc_x'], label='acc_x')
    ax3.plot(df_f_win['time'], df_f_win['acc_z'], label='acc_z')
    ax3.set_title(f"FRONT trimmed window [{start:.2f}s, {end:.2f}s]")
    ax3.set_xlabel("Time (s)")
    ax3.set_ylabel("m/s")
    ax3.grid(True)
    ax3.legend()
    fig3.suptitle("Final trimmed window (FRONT)", y=0.98)
    plt.tight_layout()

    # -------------------
    # 4) Final sway band (0.510 Hz), FRONT acc_x & acc_z
    # -------------------
    df_f_sway = df_f_win.copy()
    for axis in ("acc_x", "acc_z"):
        df_f_sway[axis] = butter_filter(df_f_sway[axis].values, fs, 0.5, btype="high")
        df_f_sway[axis] = butter_filter(df_f_sway[axis].values, fs, 10.0, btype="low")

    fig4, ax4 = plt.subplots(1, 1, figsize=(12, 4), sharex=True)
    ax4.plot(df_f_sway['time'], df_f_sway['acc_x'], label='acc_x')
    ax4.plot(df_f_sway['time'], df_f_sway['acc_z'], label='acc_z')
    ax4.set_title("FRONT sway band (0.510 Hz)")
    ax4.set_xlabel("Time (s)")
    ax4.set_ylabel("m/s")
    ax4.grid(True)
    ax4.legend()
    fig4.suptitle("Final sway band (FRONT)", y=0.98)
    plt.tight_layout()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_intermediate_csv(group, participant, trial="eyes_open1"):
    """
    Plot acc_x and acc_z from the saved intermediate CSV files 
    (front & back phones) for a given participant/trial.
    """
    base = os.path.join("..","data","intermediate", group, participant)

    front_path = os.path.join(base, f"{participant}_front_phone", f"{participant}_{trial}.csv")
    back_path  = os.path.join(base, f"{participant}_back_phone",  f"{participant}_{trial}.csv")

    df_front = pd.read_csv(front_path)
    df_back  = pd.read_csv(back_path)

    fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    axs[0].plot(df_front['time'], df_front['acc_x'], label="acc_x")
    axs[0].plot(df_front['time'], df_front['acc_z'], label="acc_z")
    axs[0].set_title(f"{participant} {trial}  FRONT phone (intermediate)")
    axs[0].set_ylabel("Acceleration (m/s)")
    axs[0].legend(); axs[0].grid(True)

    axs[1].plot(df_back['time'], df_back['acc_x'], label="acc_x")
    axs[1].plot(df_back['time'], df_back['acc_z'], label="acc_z")
    axs[1].set_title(f"{participant} {trial}  BACK phone (intermediate)")
    axs[1].set_xlabel("Time (s)")
    axs[1].set_ylabel("Acceleration (m/s)")
    axs[1].legend(); axs[1].grid(True)

    plt.tight_layout()
    plt.show()

plot_intermediate_csv("older_adults", "ann", trial="eyes_open1")
```
